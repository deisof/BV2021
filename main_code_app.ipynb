```python
N_DAYS = 1
N_SCORE = 40


class MyWidget(QMainWindow, Ui_MainWindow):
    def __init__(self):
        super().__init__()
        self.setupUi(self)
        self.initUI()

    def initUI(self):
        # self.setWindowIcon(QIcon('img_project/icon.jpg'))
        self.setFixedSize(411, 349)

        window = self.frameGeometry()
        window_central = QDesktopWidget().availableGeometry().center()
        window.moveCenter(window_central)
        self.move(window.topLeft())

        self.pushButton_2.clicked.connect(self.load)
        self.pushButton_3.clicked.connect(self.load)
        self.pushButton_4.clicked.connect(self.load)
        self.pushButton_5.clicked.connect(self.load)
        self.pushButton.clicked.connect(self.propose)
        self.pushButton_6.clicked.connect(self.change)

        nImage = QImage("img/1.png")
        sImage = nImage.scaled(QSize(411, 349))
        palette = QPalette()
        palette.setBrush(QPalette.Window, QBrush(sImage))
        self.setPalette(palette)

        self.listt = []
        self.s1, self.s2, self.s3, self.s4 = '', '', '', ''

    def change(self): 
        self.window2 = Change()
        self.window2.show()
    
    def propose(self):
        try:
            if not self.s1 or not self.s2 or not self.s3 or not self.s4:
                raise ValueError
        except ValueError:
            self.label_6.setText('Error: Заполните все поля')
            return
        self.listt.append(self.s1)
        self.listt.append(self.s2)
        self.listt.append(self.s3)
        self.listt.append(self.s4)
        self.window3 = Propose(self.listt)
        self.window3.show()
        self.hide()
    
    def load(self):
        fname = QFileDialog.getOpenFileName(self, 'Выбрать картинку', '')[0]
        need = fname.split('/')
        button = self.sender().objectName()
        if button == 'pushButton_2':
            self.label_7.setText(need[-1])
            self.s1 = fname
        elif button == 'pushButton_3':
            self.label_8.setText(need[-1])
            self.s2 = fname
        elif button == 'pushButton_4':
            self.label_9.setText(need[-1])
            self.s3 = fname
        else:
            self.label_10.setText(need[-1])
            self.s4 = fname


class Change(QMainWindow, Ui_MainWindow2):
    def __init__(self):
        super().__init__()
        self.setupUi(self)
        self.initUI()

    def initUI(self):
        # self.setWindowIcon(QIcon('img_project/icon.jpg'))
        self.setFixedSize(371, 220)

        window = self.frameGeometry()
        window_central = QDesktopWidget().availableGeometry().center()
        window.moveCenter(window_central)
        self.move(window.topLeft())

        self.pushButton.clicked.connect(self.change)

        nImage = QImage("img/1.png")
        sImage = nImage.scaled(QSize(371, 220))
        palette = QPalette()
        palette.setBrush(QPalette.Window, QBrush(sImage))
        self.setPalette(palette)

    def change(self):
        global N_DAYS
        global N_SCORES
        try:
            N_DAYS = int(self.lineEdit.text())
            N_SCORE = int(self.lineEdit_2.text())
        except Exception:
            self.label_4.setText('Error: Неправильный формат ввода')
            return
        message = QMessageBox.information(self, '', "Параметры установлены", QMessageBox.Ok)
        self.hide()


class Propose(QMainWindow, Ui_MainWindow1):
    def __init__(self, listt):
        super().__init__()
        self.setupUi(self)
        self.listt = listt
        # self.setWindowIcon(QIcon('img_project/icon.jpg'))
        self.setFixedSize(520, 533)

        window = self.frameGeometry()
        window_central = QDesktopWidget().availableGeometry().center()
        window.moveCenter(window_central)
        self.move(window.topLeft())

        self.main_propose()

        nImage = QImage("img/1.png")
        sImage = nImage.scaled(QSize(520, 533))
        palette = QPalette()
        palette.setBrush(QPalette.Window, QBrush(sImage))
        self.setPalette(palette)
    
    def main_propose(self):
        global N_DAYS
        global N_SCORE
        submissions_train = pd.read_csv(self.listt[0], delimiter=';')
        drop_cols = ['attempt_time']
        if isinstance(drop_cols, str) or (not isinstance(drop_cols, collections.Iterable)):
              drop_cols = [drop_cols]
        drop_col_names = np.intersect1d(submissions_train.columns, drop_cols)
        submissions_train.drop(drop_col_names, axis=1, inplace=True)
        submissions_train['date'] = pd.to_datetime(submissions_train.submission_time, unit='s')
        submissions_train['day'] = submissions_train.date.dt.date
        users_min_time = submissions_train.groupby('user_id', as_index=False).agg({'submission_time': 'min'}).rename(
              {'submission_time': 'min_timestamp'}, axis=1)
        users_min_time['min_timestamp'] += 60 * 60 * 24 * N_DAYS
        submission_data_d = pd.merge(submissions_train, users_min_time, how='inner', on='user_id')
        cond = submission_data_d['submission_time'] <= submission_data_d['min_timestamp']
        submission_data_d = submission_data_d[cond]
        submissions = submission_data_d
        # таблица с данными по каждому пользователю
        users_data = submissions.groupby('user_id', as_index=False).agg({'submission_time': 'max'}).rename(columns={'submission_time': 'last_timestamp'})
        # попытки сдачи практики пользователя
        users_scores = submissions.pivot_table(index='user_id',
                                                columns='status',
                                                values='step_id',
                                                aggfunc='count',
                                                fill_value=0).reset_index()
        users_data = users_data.merge(users_scores, on='user_id', how='outer')
        users_data = users_data.fillna(0)
        # колво дней на курсе
        users_days = submissions.groupby('user_id').day.nunique().to_frame().reset_index()
        users_data = users_data.merge(users_days, how='outer')
        sertificate = pd.read_csv(self.listt[2], delimiter=';')
        sertificate['certificate_url'] = pd.isna(sertificate['certificate_url'])
        users_data = users_data.merge(sertificate, on='user_id', how='outer')
        users_data = users_data[:834]
        users_data['certificate_url'] = users_data.certificate_url.fillna(False)
        users_data['certificate_url'] = users_data['certificate_url'].astype('int32')
        cheating = pd.read_csv(self.listt[3], delimiter=';')
        all_cheating = (cheating.groupby('user_id', as_index=False)['cheating_probability (%)']
                      .agg({'cheating_probability (%)': lambda a: sum(a)//len(a)})
                      .rename(columns={'step_id': cheating}))
        users_data = users_data.merge(all_cheating, on='user_id', how='outer')
        users_data = users_data.rename(columns={'cheating_probability (%)': 'cheating'})
        users_data['cheating'] = users_data.cheating.fillna(0)
        comments = pd.read_csv(self.listt[1], delimiter=';')
        comments['date'] = pd.to_datetime(comments.time_utc, unit='s')
        comments['day'] = comments.date.dt.date
        users_min_time = comments.groupby('user_id', as_index=False).agg({'time_utc': 'min'}).rename(
              {'time_utc': 'min_timestamp'}, axis=1)
        users_min_time['min_timestamp'] += 60 * 60 * 24 * N_DAYS
        comments_data_d = pd.merge(comments, users_min_time, how='inner', on='user_id')
        cond = comments_data_d['time_utc'] <= comments_data_d['min_timestamp']
        comments_data_dsubmission_data_d = comments_data_d[cond]
        all_comments = comments_data_d
        all_comments = all_comments.drop(['step_id', 'time_utc', 'date', 'day', 'min_timestamp'], axis=1)
        answer_comments = (all_comments.groupby('user_id', as_index=False)['comment_id']
                      .agg({'comments': lambda a: len(a)}))
        users_data = users_data.merge(answer_comments, on='user_id', how='outer')
        users_data['comments'] = users_data.comments.fillna(0)
        users_data = users_data[:834]
        X = users_data.drop(['user_id', 'certificate_url'], axis=1)
        y = users_data['certificate_url']
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y)
        pred_proba = joblib.load('saved_model.pkl') 
        pred_proba = pred_proba.predict(X_test)
        self.tableWidget.setRowCount(len(pred_proba))
        self.tableWidget.setColumnCount(2)
        self.tableWidget.setHorizontalHeaderLabels(['Идентификатор пользователя', 'Успешность прохождения курса'])
        ans = []
        for i in range(len(pred_proba)):
            n = (pred_proba[i] == True)
            if n:
                 n = 'Да'
            else:
                 n = 'Нет'
            ans.append([X_test.index[i], n])
        for i, elem in enumerate(ans):
            for j, val in enumerate(elem):
                self.tableWidget.setItem(i, j, QTableWidgetItem(str(val)))


if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MyWidget()
    ex.show()
    sys.exit(app.exec_())
```
